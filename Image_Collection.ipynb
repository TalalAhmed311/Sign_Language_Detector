```python

import cv2
labels = ['hello','thumbsup','thumpsdown','nice']
n_img = 20


path ='<Specify the path of dir>'
for label in labels:
    os.mkdir(os.path.join(path,label))
    webCam = cv2.VideoCapture(0)
    print(label)
    time.sleep(5)

    for n in range(n_img):
        success, img = webCam.read()
        img_name = os.path.join(path,label,label+'{}.jpg'.format(str(n)))
        cv2.imshow('img',img)
        time.sleep(2)
        cv2.imwrite(img_name, img)
        print('done'+' '+str(n))
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('e'):
            break
    webCam.release()
```
